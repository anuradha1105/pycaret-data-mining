```python
import pandas as pd
from pycaret.classification import *

df = pd.read_csv("iris.csv")
print(df.shape)
print(df.head())
```

```python
s = setup(
    data=df,
    target="species",
    session_id=123,
    normalize=True,
    use_gpu=False,
)

best = compare_models()
plot_model(best, plot="confusion_matrix")
save_model(best, "multiclass_iris_model")
```
